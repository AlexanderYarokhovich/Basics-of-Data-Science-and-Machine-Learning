Теперь проведем анализ выбросов. Начнем с числовых признаков, таких как AnnualIncome, SpentOnWines, SpentOnFruits, SpentOnMeat, SpentOnFish, SpentOnSweets, SpentOnGold, а также другие показатели, связанные с тратами и покупками. Используем метод межквартильного размаха (IQR) для выявления выбросов.

```python
import numpy as np

# Функция для выявления выбросов методом IQR
def detect_outliers_iqr(data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = data[(data[column] < lower_bound) | (data[column] > upper_bound)]
    return outliers.shape[0]

# Выбираем числовые признаки для анализа
numeric_columns = ['AnnualIncome', 'SpentOnWines', 'SpentOnFruits', 'SpentOnMeat',
                   'SpentOnFish', 'SpentOnSweets', 'SpentOnGold', 'DaysSinceLastPurchase',
                   'DiscountedPurchases', 'OnlinePurchases', 'CatalogPurchases',
                   'InStorePurchases', 'WebsiteVisitsLastMonth', 'RevenueGenerated']

# Проверяем количество выбросов по каждому столбцу
outliers_count = {col: detect_outliers_iqr(df, col) for col in numeric_columns}
print("\nКоличество выбросов в каждом числовом столбце:")
print(outliers_count)

```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Строим boxplot и histogram для выявления выбросов
def plot_feature_distribution(data, column):
    fig, axes = plt.subplots(1, 2, figsize=(12, 5))

    # Boxplot
    sns.boxplot(x=data[column], ax=axes[0])
    axes[0].set_title(f'Boxplot: {column}')

    # Гистограмма
    sns.histplot(data[column], bins=30, kde=True, ax=axes[1])
    axes[1].set_title(f'Histogram: {column}')

    plt.tight_layout()
    plt.show()

# Анализируем признаки с выбросами
for col in ['AnnualIncome', 'SpentOnWines', 'SpentOnFruits', 'SpentOnMeat',
            'SpentOnFish', 'SpentOnSweets', 'SpentOnGold', 'DiscountedPurchases',
            'OnlinePurchases', 'CatalogPurchases', 'WebsiteVisitsLastMonth']:
    plot_feature_distribution(df, col)
```

```python
# Функция для обработки выбросов
def handle_outliers(df, col, method="iqr"):
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    if method == "iqr":
        df[col] = df[col].clip(lower=lower_bound, upper=upper_bound)
    elif method == "percentile":
        upper_limit = df[col].quantile(0.99)
        df[col] = df[col].apply(lambda x: upper_limit if x > upper_limit else x)

# Обрабатываем выбросы
for col in numeric_columns:
    num_outliers = (df[col] > df[col].quantile(0.99)).sum()
    total_count = df.shape[0]
    outlier_ratio = num_outliers / total_count

    if outlier_ratio < 0.1:
        handle_outliers(df, col, method="percentile")
    else:
        handle_outliers(df, col, method="iqr")

# Проверяем количество оставшихся выбросов
outliers_after = {col: ((df[col] > df[col].quantile(0.99)).sum()) for col in numeric_columns}
print("\nВыбросы после обработки (должны быть значительно уменьшены):")
print(outliers_after)
```

```python
from sklearn.cluster import KMeans, DBSCAN, AgglomerativeClustering


# Выбираем только числовые признаки для кластеризации
df_cluster_numeric = df_cluster.select_dtypes(include=['int64', 'float64'])

# 1. Метод K-Means (зададим 4 кластера)
kmeans = KMeans(n_clusters=4, random_state=42, n_init=10)
df_cluster['KMeans_Cluster'] = kmeans.fit_predict(df_cluster_numeric)

# 2. Иерархическая (агломеративная) кластеризация
agg_clustering = AgglomerativeClustering(n_clusters=4)
df_cluster['Agglomerative_Cluster'] = agg_clustering.fit_predict(df_cluster_numeric)

# 3. DBSCAN (плотностной метод)
dbscan = DBSCAN(eps=0.3, min_samples=5)
df_cluster['DBSCAN_Cluster'] = dbscan.fit_predict(df_cluster_numeric)

# Визуализируем распределение кластеров для KMeans
plt.figure(figsize=(8, 5))
sns.countplot(data=df_cluster, x="KMeans_Cluster", hue="KMeans_Cluster", palette="viridis", legend=False)
plt.title("Распределение клиентов по кластерам (K-Means)")
plt.xlabel("Кластер")
plt.ylabel("Количество клиентов")
plt.show()

# Проверяем результаты кластеризации
df_cluster[['KMeans_Cluster', 'Agglomerative_Cluster', 'DBSCAN_Cluster']].head()
```

### 2.2. Определить оптимальное количество кластеров.

```python
# Определение оптимального количества кластеров с помощью метода локтя и коэффициента силуэта
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt

# Диапазон количества кластеров для тестирования
cluster_range = range(2, 11)

# Списки для оценки качества кластеризации
inertia_values = []  # Метод "локтя"
silhouette_scores = []  # Коэффициент силуэта

for k in cluster_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    cluster_labels = kmeans.fit_predict(df_cluster_numeric)
    inertia_values.append(kmeans.inertia_)
    silhouette_scores.append(silhouette_score(df_cluster_numeric, cluster_labels))

# Визуализация метода локтя
plt.figure(figsize=(10, 4))
plt.plot(cluster_range, inertia_values, marker='o', linestyle='-')
plt.xlabel("Количество кластеров")
plt.ylabel("Inertia (Сумма квадратов расстояний до центроидов)")
plt.title("Метод локтя для определения количества кластеров")
plt.show()

# Визуализация коэффициента силуэта
plt.figure(figsize=(10, 4))
plt.plot(cluster_range, silhouette_scores, marker='o', linestyle='-')
plt.xlabel("Количество кластеров")
plt.ylabel("Средний коэффициент силуэта")
plt.title("Оптимальное количество кластеров по силуэту")
plt.show()
```

Метод локтя:
    Резкое падение до 4-5 кластеров, затем снижение становится плавным.
    Значит, оптимальное количество кластеров около 4-6, после этого добавление кластеров мало что меняет.
Коэффициент силуэта:
    Максимум при 2-4 кластерах, затем снижение.
    Оптимальный баланс 4-5 кластеров, так как дальше качество падает.

### 2.3. Оцените качество полученной кластеризации.

```python
from sklearn.metrics import silhouette_score, davies_bouldin_score, calinski_harabasz_score
import numpy as np

# Функция для вычисления метрик качества кластеризации
def evaluate_clustering(data, labels, method_name):
    # Проверяем, есть ли менее 2 кластеров (например, DBSCAN может давать -1, означающий "шум")
    unique_clusters = len(np.unique(labels))
    if unique_clusters < 2:
        print(f"{method_name}: Недостаточно кластеров для вычисления метрик.\n")
        return
    
    # Вычисляем метрики
    silhouette_avg = silhouette_score(data, labels)
    davies_bouldin = davies_bouldin_score(data, labels)
    calinski_harabasz = calinski_harabasz_score(data, labels)
    
    # Вывод результатов
    print(f"Метрики для {method_name}:")
    print(f"- Коэффициент силуэта: {silhouette_avg:.4f} (чем выше, тем лучше)")
    print(f"- Индекс Дэвиса-Болдина: {davies_bouldin:.4f} (чем ниже, тем лучше)")
    print(f"- Индекс Калински-Харабаса: {calinski_harabasz:.4f} (чем выше, тем лучше)\n")

# Оценка кластеризации для каждого метода
evaluate_clustering(df_cluster_numeric, df_cluster['KMeans_Cluster'], "K-Means")
evaluate_clustering(df_cluster_numeric, df_cluster['Agglomerative_Cluster'], "Agglomerative Clustering")

# Проверка, есть ли -1 (шум) в DBSCAN
if len(np.unique(df_cluster['DBSCAN_Cluster'])) > 1:
    evaluate_clustering(df_cluster_numeric, df_cluster['DBSCAN_Cluster'], "DBSCAN")
else:
    print("DBSCAN: Недостаточно кластеров для оценки качества.\n")

```

### 2.4. Проведите кластеризацию наилучшим образом и сохраните результат кластеризации в новый столбец датасета.

```python
from sklearn.decomposition import PCA
# 7. Определяем оптимальное количество главных компонент для PCA
pca = PCA()
pca.fit(df_cluster)
explained_variance = np.cumsum(pca.explained_variance_ratio_)

# Визуализация накопленного объясненного дисперсионного вклада
plt.figure(figsize=(10, 4))
plt.plot(range(1, len(explained_variance) + 1), explained_variance, marker='o', linestyle='-')
plt.xlabel("Количество главных компонент")
plt.ylabel("Объясненная дисперсия")
plt.title("Выбор количества компонент PCA")
plt.axhline(y=0.85, color='r', linestyle='--', label="90% дисперсии")
plt.legend()
plt.show()

# 8. Применяем PCA с оптимальным количеством компонент (выберем 85% дисперсии)
optimal_components = np.argmax(explained_variance >= 0.85) + 1
pca = PCA(n_components=optimal_components)
df_pca = pca.fit_transform(df_cluster)

# 9. Кластеризация K-Means с 5 кластерами
kmeans_final = KMeans(n_clusters=5, random_state=42, n_init=10)
df_cluster['Final_KMeans_Cluster'] = kmeans_final.fit_predict(df_pca)

# 10. Визуализация распределения клиентов по кластерам
plt.figure(figsize=(8, 5))
sns.countplot(data=df_cluster, x="Final_KMeans_Cluster", hue="Final_KMeans_Cluster", palette="viridis", legend=False)
plt.title("Распределение клиентов по кластерам (K-Means, PCA)")
plt.xlabel("Кластер")
plt.ylabel("Количество клиентов")
plt.show()

# 11. Функция для оценки качества кластеризации
def evaluate_clustering(data, labels, method_name):
    unique_clusters = len(np.unique(labels))
    if unique_clusters < 2:
        print(f"{method_name}: Недостаточно кластеров для вычисления метрик.\n")
        return
    
    silhouette_avg = silhouette_score(data, labels)
    davies_bouldin = davies_bouldin_score(data, labels)
    calinski_harabasz = calinski_harabasz_score(data, labels)
    
    print(f"Метрики для {method_name}:")
    print(f"- Коэффициент силуэта: {silhouette_avg:.4f} (чем выше, тем лучше)")
    print(f"- Индекс Дэвиса-Болдина: {davies_bouldin:.4f} (чем ниже, тем лучше)")
    print(f"- Индекс Калински-Харабаса: {calinski_harabasz:.4f} (чем выше, тем лучше)\n")

# 12. Оценка качества финальной кластеризации
evaluate_clustering(df_pca, df_cluster['Final_KMeans_Cluster'], "K-Means + PCA")

# 13. Вывод первых строк с кластерами
from IPython.display import display
display(df_cluster[['Final_KMeans_Cluster']].head())
columns_to_remove = ['KMeans_Cluster', 'Agglomerative_Cluster', 'DBSCAN_Cluster']
df_cluster.drop(columns=columns_to_remove, inplace=True, errors='ignore')
```

**Профиль:** Самый низкий доход, минимальные траты, не реагируют на рекламу.  
**Предложение:**  
- Бюджетные продуктовые наборы.  
- Акции «Купи больше – плати меньше».  
- Бесплатные купоны для повышения интереса к покупкам.

---

#### **Кластер 2: Семейные клиенты с хорошим доходом, но низкой вовлеченностью**
**Профиль:** Хороший доход, много детей, активность на сайте, редкие покупки.  
**Предложение:**  
- Семейные скидки на детское питание и бытовые товары.  
- Персональные рекомендации на основе веб-активности.  
- Программы подписки на регулярные поставки товаров.

---

#### **Кластер 3: Высокий доход, активные покупатели**
**Профиль:** Максимальные траты, лояльные клиенты, активные в покупках.  
**Предложение:**  
- VIP-программы с эксклюзивными скидками.  
- Персонализированные предложения на премиальные товары.  
- Ранний доступ к новым коллекциям и продуктам.

---

#### **Кластер 4: Высокий доход, молодые покупатели**
**Профиль:** Высокий доход, много одиночек, средние траты, любят акции.  
**Предложение:**  
- Скидки на премиальные товары (кофе, вино, аксессуары).  
- Персональные кэшбэки за активные покупки.  
- Развлекательные мероприятия (дегустации, мастер-классы).

---

### **Общие рекомендации**
- Разработать индивидуальные промо-кампании для каждого сегмента.  
- Использовать автоматизированные персональные предложения на основе покупательской активности.  
- Применять кросс-продажи предлагать товары, дополняющие уже купленные.  


### **4.2. Как рекомендации улучшат маркетинговую стратегию компании?**

Внедрение персонализированных предложений для каждого кластера клиентов позволит компании **оптимизировать маркетинговую стратегию**, сократить затраты на рекламу и повысить эффективность кампаний. Вот ключевые выгоды:

---

### **1. Рост конверсии и удержание клиентов**  
Персонализированные предложения повышают вероятность покупки, так как клиенты получают релевантные скидки и акции.  
Программы лояльности увеличивают удержание клиентов, побуждая их совершать повторные покупки.  

**Ожидаемый результат:**  
- Увеличение среднего чека.  
- Снижение оттока клиентов.  

---

### **2. Оптимизация рекламного бюджета**  
Фокусирование на целевых сегментах снижает неэффективные затраты на массовую рекламу.  
Использование данных о покупательском поведении позволяет инвестировать в наиболее перспективные аудитории.  

**Ожидаемый результат:**  
- Снижение затрат на рекламу.  
- Увеличение ROI от маркетинговых кампаний.  

---

### **3. Развитие стратегий кросс-продаж**  
Для активных покупателей (кластер 3 и 4) можно предлагать дополнительные товары (например, вино и сыры, кофе и десерты).  
Для семейных клиентов (кластер 0 и 2) эффективны комплектные предложения (детское питание + бытовые товары).  

**Ожидаемый результат:**  
- Увеличение выручки за счёт сопутствующих продаж.  
- Формирование новых привычек потребления у клиентов.  

---

### **4. Повышение вовлеченности через цифровые каналы**  
Кластеры с высокой активностью в интернете могут получать персонализированные e-mail и push-уведомления.  
Развитие программ подписок и специальных предложений увеличивает кликабельность и отклик на маркетинговые кампании.  

**Ожидаемый результат:**  
- Повышение открываемости писем и отклика на предложения.  
- Увеличение вовлечённости клиентов в цифровые каналы.  

---

### **5. Укрепление бренда и репутации**  
Персонализация маркетинга формирует у клиентов ощущение заботы со стороны компании.  
Индивидуальный подход улучшает восприятие бренда, повышая доверие и лояльность клиентов.  

**Ожидаемый результат:**  
- Улучшение репутации компании.  
- Рост количества рекомендаций и «сарафанного радио».  

---

### **Вывод:**  
Реализация предложенных рекомендаций позволит сделать маркетинг более целевым и эффективным, увеличив доходность каждого клиента и снизив затраты на привлечение новых покупателей. В результате компания получит устойчивый рост продаж, увеличение лояльности клиентов и укрепление позиций на рынке.
