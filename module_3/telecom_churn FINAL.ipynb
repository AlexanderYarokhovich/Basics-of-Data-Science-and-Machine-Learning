```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder

data = pd.read_csv("dataset/telcom_customer_churn.csv") 

data_info = data.info()
data_head = data.head()
data_missing = data.isnull().sum()

data_info, data_head, data_missing
```

```python
# Расчёт процента пропусков для каждого столбца
missing_percentage = data.isnull().mean() * 100

# Разделение столбцов по проценту пропусков
low_missing = missing_percentage[missing_percentage <= 10].index  # <=10%
moderate_missing = missing_percentage[(missing_percentage > 10) & (missing_percentage <= 40)].index  # 10-40%
high_missing = missing_percentage[missing_percentage > 40].index  # >40%

# Заполнение пропусков для столбцов с низким процентом пропусков
for col in low_missing:
    if data[col].dtype == 'object':
        data[col] = data[col].fillna(data[col].mode()[0])
    else:
        data[col] = data[col].fillna(data[col].mean())

# Удаление столбцов с высоким процентом пропусков
data = data.drop(columns=high_missing)

# Удаление столбца 'Customer_ID'
if 'Customer_ID' in data.columns:
    data = data.drop(columns=['Customer_ID'])

# Обработка столбцов с умеренным процентом пропусков (10-40%),
# выбираем вариант заполнения, который дал лучшие результаты.

# Заполнение 'lor' медианным значением (для числового признака)
data['lor'] = data['lor'].fillna(data['lor'].median())

# Заполнение 'adults' модальным значением (наиболее часто встречающееся значение)
data['adults'] = data['adults'].fillna(data['adults'].mode()[0])

# Заполнение 'income':
# Если присутствует признак 'age', группируем по возрасту и заполняем медианой в каждой группе.
if 'age' in data.columns:
    data['income'] = data.groupby('age')['income'].transform(lambda x: x.fillna(x.median()))
else:
    # Если 'age' отсутствует, заполняем общей медианой.
    data['income'] = data['income'].fillna(data['income'].median())
# Данные где пропусков от 10 до 40 процентов обрабатывал отдельно, так как думал что, перед обучением модели буду по разному их обрадатывать, 
# и смотреть, что даст лучший результат, по факту, разницы почти нет отсановился на следующем варианте:
# Заполнение пропусков
```

```python
# В этом блоке кода я провожу анализ категориальных переменных в датасете.
# Сначала я определяю все столбцы, содержащие категориальные данные (тип "object"),
# затем вывожу их уникальные значения для предварительного анализа.
# Далее, в зависимости от количества уникальных значений, я разделяю столбцы на две группы:
# - Для столбцов с небольшим количеством уникальных значений (<=10) применяю one-hot кодирование.
# - Для столбцов с большим количеством уникальных значений (>10) применяю label encoding.

# Определяем категориальные переменные
categorical_columns = data.select_dtypes(include=['object']).columns

# Вывод уникальных значений для каждой категориальной переменной
categorical_summary = {col: data[col].unique() for col in categorical_columns}

# Разделяем столбцы на группы для кодирования:
# one_hot_columns - для переменных с небольшим числом уникальных значений (<=10)
# label_encode_columns - для переменных с большим числом уникальных значений (>10)
one_hot_columns = [col for col in categorical_columns if data[col].nunique() <= 10]
label_encode_columns = [col for col in categorical_columns if data[col].nunique() > 10]

# Применяем one-hot кодирование для столбцов с небольшим количеством уникальных значений
data = pd.get_dummies(data, columns=one_hot_columns, drop_first=True)

# Применяем label encoding для столбцов с большим количеством уникальных значений
label_encoders = {}
for col in label_encode_columns:
    le = LabelEncoder()
    data[col] = le.fit_transform(data[col])
    label_encoders[col] = le

# Вывод информации о преобразованных данных
processed_data_info = data.info()
processed_data_head = data.head()

# Проверка очищенных данных на наличие пропусков
data_cleaned_missing = data.isnull().sum()

print(data_cleaned_missing)
```

На основе анализа выше, понял что есть много сильно корелированных между собой признаков, и конечно удаляю, так же видно много выбросов, 
их тоже решил обработать. и для обучения буду использовать только топ самых полезных признаков, так как очень долго модель будет учится.

```python
from sklearn.ensemble import RandomForestClassifier

# 1) Оценка важности признаков на исходных данных
X = data.drop(columns=['churn'])
y = data['churn']

rf = RandomForestClassifier(random_state=42)
rf.fit(X, y)

feature_importances = pd.Series(rf.feature_importances_, index=X.columns)

# 2) Удаление сильно коррелированных признаков, оставляем более важный
correlation_threshold = 0.9
corr_matrix = X.corr().abs()
upper_triangle = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(bool))

features_to_drop = set()

for col in upper_triangle.columns:
    for row in upper_triangle.index:
        val = upper_triangle.loc[row, col]
        if pd.isna(val):
            continue
        
        if val > correlation_threshold:
            # Проверяем, не удаляли ли уже
            if row in features_to_drop or col in features_to_drop:
                continue
            
            # Сравниваем важность (feature_importances). Удаляем менее важный
            if feature_importances[row] >= feature_importances[col]:
                features_to_drop.add(col)
            else:
                features_to_drop.add(row)

print("Удаляем за сильную корреляцию:", features_to_drop)

# Удаляем из X из основного data
X = X.drop(columns=list(features_to_drop))
data = pd.concat([X, y], axis=1)

# 3) «Прижимание» (capping) выбросов в числовых признаках методом IQR
def cap_outliers_iqr(df, column, threshold=3.0):
    """
    Заменяет слишком большие/маленькие значения на границы по IQR.
    threshold=3.0 - более мягкий отсек, чем классический 1.5. так как прочитал, что в телеком данных большие выбросы норма
    """
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    
    lower_fence = Q1 - threshold * IQR
    upper_fence = Q3 + threshold * IQR
    
    df.loc[df[column] < lower_fence, column] = lower_fence
    df.loc[df[column] > upper_fence, column] = upper_fence

num_cols = X.select_dtypes(include=['int64','float64']).columns
for col in num_cols:
    cap_outliers_iqr(X, col, threshold=3.0)

# Обновляем data (теперь с капированными выбросами)
data = pd.concat([X, y], axis=1)

# 4) Снова оцениваем важность признаков и берём топ-50
X_after = data.drop(columns=['churn'])
y_after = data['churn']

rf2 = RandomForestClassifier(random_state=42)
rf2.fit(X_after, y_after)

feature_importances2 = pd.Series(rf2.feature_importances_, index=X_after.columns)
feature_importances2 = feature_importances2.sort_values(ascending=False)

top_n = 50
top_features = feature_importances2.head(top_n).index

X_top = X_after[top_features]

print("Топ признаков по важности (после корреляции/выбросов):")
print(feature_importances2.head(top_n))
print("\nИтоговые размеры X_top:", X_top.shape)
```

## Этап 4
#### Тонкая настройка модели:
+ Использовал методы гиперпараметрической оптимизации, такие как Randomized Search, для улучшения качества модели.
+ Применил метод балансировки классов (с помощью SMOTE ).

```python
from sklearn.model_selection import train_test_split, RandomizedSearchCV
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score
from imblearn.over_sampling import SMOTE

# Балансировка SMOTE
smote = SMOTE(random_state=42)
X_train_balanced, y_train_balanced = smote.fit_resample(X_train_scaled, y_train)

# ----- Randomized Search для Gradient Boosting -----
gb = GradientBoostingClassifier(random_state=42)

# Набор параметров
param_dist = {
    'n_estimators': [100, 150, 200, 300],
    'learning_rate': [0.01, 0.05, 0.1, 0.2],
    'max_depth': [3, 4, 5, 6],
    'min_samples_split': [2, 5, 10],
    'subsample': [0.8, 0.9, 1.0]
}

# n_iter=20 — переберёт 20 случайных комбинаций из заданного пространства
# cv=5  — кросс-валидация в 3 фолда, scoring='roc_auc'
# Конечно чем выше n_iter и cv тем выше качество, но мой ноут не тянет, и поэтому выбран понижение 

random_search = RandomizedSearchCV(
    estimator=gb,
    param_distributions=param_dist,
    n_iter=10,
    scoring='roc_auc',
    cv=2,
    random_state=42,
    n_jobs=-1,
    verbose=1  
)

# Запуск поиска
random_search.fit(X_train_balanced, y_train_balanced)

# Лучшая модель
best_gb = random_search.best_estimator_
print("Лучшие параметры (Gradient Boosting):", random_search.best_params_)
print("Лучший CV ROC-AUC:", random_search.best_score_)

# Оценка на тестовой выборке
y_pred_gb = best_gb.predict(X_test_scaled)
y_proba_gb = best_gb.predict_proba(X_test_scaled)[:, 1]

gb_results = {
    "Model": "Gradient Boosting (RandomSearch)",
    "CV ROC-AUC": random_search.best_score_,
    "Accuracy": accuracy_score(y_test, y_pred_gb),
    "Precision": precision_score(y_test, y_pred_gb),
    "Recall": recall_score(y_test, y_pred_gb),
    "F1-Score": f1_score(y_test, y_pred_gb),
    "ROC-AUC": roc_auc_score(y_test, y_proba_gb)
}
results_df = pd.DataFrame([gb_results])
print(results_df)

```
